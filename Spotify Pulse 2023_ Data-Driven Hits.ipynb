DATA CLEANING

```python
# 1. Clean 'streams' column

# Find non-numeric entries

non_numeric_streams = df[pd.to_numeric(df['streams'], errors='coerce').isna()]['streams']
print("Non-numeric stream values found:", non_numeric_streams.tolist())
```

```python
# Convert streams to numeric, forcing errors to NaN then dropping or filling

df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
```

```python
# 2. Clean 'in_deezer_playlists' and 'in_shazam_charts'

# These are objects likely due to commas

df['in_deezer_playlists'] = df['in_deezer_playlists'].str.replace(',', '')
df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')

df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',', '')
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')
```

```python
# 3. Handle Missing Values

# key: fill with 'Unknown'
df['key'] = df['key'].fillna('Unknown')

# in_shazam_charts: fill with 0 (assuming not on chart)
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(0)

# streams: drop the row if stream is NaN (since it's the target variable usually)
df = df.dropna(subset=['streams'])
```

```python
# 4. Check for duplicates

duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
df = df.drop_duplicates()
```

```python
# Final Check

print("\nCleaned Data Info:")
print(df.info())
```

```python
# Save cleaned data

df.to_csv('spotify_2023_cleaned.csv', index=False)
```

# Phase 3 - Findings

1. **The Power of "Reachability" (Platform Influence)**
- The Playlist Engine: Inclusion in Spotify playlists is the single most dominant factor affecting a song's popularity. There is a nearly linear relationship between the number of playlist adds and total streams
- Cross-Platform Synergy: Success is rarely isolated to one app. A strong "halo effect" exists where tracks that perform well on Spotify almost always see high placement on Apple Music and Deezer charts, amplifying their global reach
- Discovery Tools: While Spotify is the volume leader, platforms like Shazam act as "early warning" systems for viral hits, showing that "reachability" starts with curiosity and discovery.
2. **Musical "Hit" Characteristics**
- The Tempo Sweet Spot: Most chart-toppers are clustered in the 100–140 BPM range. This tempo is versatile enough for both active listening and background streaming, making it the industry standard for 2023.
- Vibe over Math: Statistical correlation shows that no single audio feature (like $Danceability\%$ or $Energy\%$) guarantees a hit. Success is holistic; while high energy and danceability are common, they must work in tandem with marketing and artist brand.
- Mode Neutrality: Although Major keys (associated with happy/bright tones) are more frequent, Minor key songs (associated with moody/darker tones) achieve equal streaming success on a per-song basis.
3. **Release Strategy & Temporal Trends**
- The "Golden Windows": Release timing is highly strategic. January and May are the peak months for successful launches, allowing songs to either capture the "New Year" listener reset or build momentum for the "Song of the Summer" title.
- Accumulation Effect: Songs released earlier in the calendar year have a significantly higher average stream count in the 2023 snapshot, proving that "staying power" and time-on-platform are crucial for reaching the billion-stream milestone.
4. **Artist & Collaboration Dynamics**
- Solo vs. Collaborative: Solo artists and duets (1–2 artists) consistently outperform larger collaborations. Data suggests that listeners prefer a clear artist identity, and too many contributors can sometimes dilute a track's appeal.
- The "Superstar" Skew: A small group of elite artists (e.g., Taylor Swift, Bad Bunny, Drake) accounts for a disproportionate share of total streams. Their brand "reachability" is so high that their tracks often bypass the traditional requirements for specific audio features.
5. **Statistical Profile of a 2023 Hit**
- Typical Audio Profile: High Danceability ($\approx 67\%$), High Energy ($\approx 64\%$), and very low Instrumentalness ($<2\%$).
- The Popularity Gap: The difference between a "Top 10" track and an "Average" track in the dataset is massive, with the top 1% of songs having streams that are 10x to 20x higher than the median, highlighting a "winner-takes-all" streaming economy.

# Phase 4 - Summary, Key Insights and Recomended strategy

**Summary**

The analysis of Spotify’s 2023 most-streamed songs reveals that global popularity is no longer driven by a single "hit formula" or specific musical attribute. Instead, success is the result of a multi-platform ecosystem where reachability—defined by playlist inclusions and cross-app presence—is the primary engine for growth. While technical features like high energy and danceability remain the standard for the top charts, they act more as "entry requirements" rather than guarantees of success. Ultimately, the data shows that a song’s lifecycle is heavily influenced by strategic release timing and the ability to bridge the gap between different streaming services.

**Key-Insights**

*Playlisting is the Kingmaker* : The strongest statistical relationship in the dataset is between playlist inclusions and total streams. Songs in the top tier are almost always characterized by massive Spotify and Apple Music playlist counts, proving that curated discovery is more powerful than organic search.

*The "Vibe" Over Key/Mode* : Listeners in 2023 are "mode-agnostic." While Major keys are more common, Minor keys achieve nearly identical median stream counts. This indicates that the "vibe" or emotional resonance of a song is more important to the audience than the technical musical scale.

*Collaboration Paradox* : While collaborations are popular for marketing, solo tracks and duets (1–2 artists) actually maintain higher average stream volumes. This suggests that a clear, focused artist brand often resonates more deeply with the streaming audience than overcrowded "feature" tracks.

*The Timing Advantage* : Tracks released in January and May show a distinct advantage in accumulating long-term reach. Early-year releases benefit from a full calendar year of data accumulation, while May releases perfectly capture the "Summer Hit" momentum that carries through the Q3 peak.

**Recommendations & Strategy**

Based on the findings, the following strategies are recommended for artists, labels, and digital marketers:

A. *The "Reachability" Strategy*

•	 Prioritize Third-Party Playlisting: Since playlist count is the leading indicator of streams, the primary marketing budget should be allocated to playlist pitching (Spotify Editorial, Apple Music Essentials, and high-traffic independent curators) rather than just traditional social media ads.

•	Cross-Platform Syncing: Ensure a song is optimized for Shazam discovery (clean metadata and early teaser clips). Songs that trend on Shazam charts often see a direct "pull-through" effect to Spotify playlists.

B. *The Audio Optimization Strategy*

•	Target the "Dance-Energy" Quadrant: For a track aimed at the Top 40, aim for a BPM between 110–130, Energy > 60%, and Danceability > 65%. This technical profile fits the most successful clusters identified in the multivariate analysis.

•	Avoid "Feature Fatigue": Limit collaborations to one guest artist (duet) to maintain brand clarity and ensure the lead artist's core fanbase remains the primary driver of streams.

C. *The Temporal Launch Strategy*

•	Strategic Release Windows: Schedule "Priority 1" releases for early Q2 (April/May) to capitalize on the higher average stream trends found in the summer months.

•	Longevity over Virality: Focus on "sticky" audio features (lower speechiness, higher production value) to ensure the song stays on playlists for 6+ months, as the top-performing tracks in 2023 were those that maintained consistent play rates over long periods rather than one-week spikes.
